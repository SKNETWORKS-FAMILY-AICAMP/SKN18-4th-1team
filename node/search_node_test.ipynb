# 설계

## Search_node
- 사용자의 질문을 청크 5개로 나누는 노드  
- RAG를 사용함  
- Classify_node에서 의료질문으로 판별된 것들이 Search_node로 넘어옴

## 개발 흐름  
1. 사용자로부터 질문을 입력 받는다.
2. 질병 DataSet에서 RAG해서 사용자의 질문을 받아온다.
3. 사용자의 질문을 textsplitter로 나눈다.
4. 나눈 텍스트를 임베딩해준다.
5. 청크 5개로 정리해준다.
